```python
# prompt: bill of materials

import pandas as pd
# Sample data
data = {'Part': ['A', 'B', 'C', 'D'],
        'Quantity': [2, 3, 1, 4],
        'Unit Cost': [10.5, 5.2, 12.0, 8.75]}

df = pd.DataFrame(data)

# Calculate total cost for each part
df['Total Cost'] = df['Quantity'] * df['Unit Cost']

# Display the bill of materials
print(df)

# Calculate and print the grand total
grand_total = df['Total Cost'].sum()
print("\nGrand Total:", grand_total)
```

```python
# prompt: visualize

import matplotlib.pyplot as plt

# Bar chart for quantity of each part
plt.figure(figsize=(8, 6))
plt.bar(df['Part'], df['Quantity'])
plt.xlabel('Part')
plt.ylabel('Quantity')
plt.title('Quantity of Each Part')
plt.show()

# Pie chart for total cost distribution
plt.figure(figsize=(8, 6))
plt.pie(df['Total Cost'], labels=df['Part'], autopct='%1.1f%%')
plt.title('Total Cost Distribution')
plt.show()
```
