```python
def where_used(assy, lpn):
    # Find the index of the specified part number and add it to a list
    indices = assy[assy['pn'] == lpn].index.tolist()
    # create a list for where used assemblies
    where_used_assemblies = []
    for index in indices:
        current_level = assy.at[index, 'level']
        for i in range(index -1, -1, -1):
            if current_level == assy.at[i, 'level']:
                continue
            elif current_level == assy.at[i, 'level'] + 1:
                where_used_assemblies.append(assy.at[i, 'pn'])
                current_level -= 1
            if current_level == 0:
                break
    return where_used_assemblies
    
print(where_used(assy, "10-0021"))
```
